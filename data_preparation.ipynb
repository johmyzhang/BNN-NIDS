```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/Wednesday-workingHours.pcap_ISCX.csv')
df
```

```python
selected_headers = [
    ' Flow Duration',
    'Total Length of Fwd Packets',  # Total Len F.Packets
    ' Total Length of Bwd Packets',  # Total Len B.Packets
    ' Fwd Packet Length Min',  # F.Packet Len Min
    ' Fwd Packet Length Mean',  # F.Packet Len Mean
    ' Bwd Packet Length Min',  # B.Packet Len Min
    ' Bwd Packet Length Std',  # B.Packet Len Std
    ' Flow IAT Mean',
    ' Flow IAT Std',
    ' Flow IAT Min',
    ' Fwd IAT Mean',  # FIAT Mean
    ' Fwd IAT Min',  # FIAT Min / Fwd IAT Min
    ' Bwd IAT Mean',
    ' Bwd IAT Min',
    'Fwd PSH Flags',  # FPSH Flags
    ' PSH Flag Count',
    ' ACK Flag Count',
    ' SYN Flag Count',
    'Fwd Packets/s',  # F.Packets/s
    ' Bwd Packets/s',  # B.Packets/s
    ' Subflow Fwd Bytes',  # Subflow F.Bytes
    ' Subflow Bwd Bytes',  # Subflow B.Bytes
    'Init_Win_bytes_forward',  # Init Win F.Bytes
    ' Init_Win_bytes_backward',  # Init Win B.Bytes
    'Active Mean',
    ' Active Min',
    ' Label'  # Assuming you need the label column for classification
]
```

```python
df_selected = df[selected_headers]
df_selected['Fwd Packets/s'] = df_selected['Fwd Packets/s'].astype(int)
df_selected[' Bwd Packets/s'] = df_selected[' Bwd Packets/s'].astype(int)
df_selected[' Bwd IAT Mean'] = df_selected[' Bwd IAT Mean'].astype(int)
df_selected[' Fwd IAT Mean'] = df_selected[' Fwd IAT Mean'].astype(int)
df_selected[' Flow IAT Std'] = df_selected[' Flow IAT Std'].astype(int)
df_selected[' Flow IAT Mean'] = df_selected[' Flow IAT Mean'].astype(int)
df_selected['Active Mean'] = df_selected['Active Mean'].astype(int)
df_selected[' Bwd Packet Length Std'] = df_selected[' Bwd Packet Length Std'].astype(int)
df_selected[' Fwd Packet Length Mean'] = df_selected[' Fwd Packet Length Mean'].astype(int)

df_selected
```

```python
df_selected[' Label'].value_counts()
```

```python
from sklearn.preprocessing import QuantileTransformer

# Transform all features into the scale of [0,1]
numeric_features = df_selected.dtypes[df_selected.dtypes != 'object'].index
scaler = QuantileTransformer()
df_selected[numeric_features] = scaler.fit_transform(df_selected[numeric_features])
```

```python
df_selected.describe()
```

```python
# Multiply the feature values by 255 to transform them into the scale of [0,255]
df_selected[numeric_features] = df_selected[numeric_features].apply(
    lambda x: (x*255).astype(int))
```

```python
df_selected.describe()
mapping = {'DoS Hulk': 1, 'BENIGN': 0, 'DoS GoldenEye': 2, 'DoS slowloris': 3, 'DoS Slowhttptest': 4, 'Heartbleed': 5}
df_selected[' Label'] = df_selected[' Label'].replace(mapping)
```

```python
df_selected.describe()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df_selected, test_size=0.2, random_state=42)

train.to_csv('data/train_ddos_2.csv', index=False, header=False)
test.to_csv('data/test_ddos_2.csv', index=False, header=False)
```
