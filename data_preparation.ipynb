```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.read_csv('data/Friday-WorkingHours-Morning.pcap_ISCX.csv')
df2 = pd.read_csv('data/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv')
df3 = pd.read_csv('data/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv')
df4 = pd.read_csv('data/Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv')
df5 = pd.read_csv('data/Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv')
df6 = pd.read_csv('data/Tuesday-WorkingHours.pcap_ISCX.csv')
df7 = pd.read_csv('data/Wednesday-WorkingHours.pcap_ISCX.csv')
df8 = pd.read_csv('data/Monday-WorkingHours.pcap_ISCX.csv')
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], ignore_index=True)
```

```python
selected_headers = [
    ' Flow Duration',
    'Total Length of Fwd Packets',  # Total Len F.Packets
    ' Total Length of Bwd Packets',  # Total Len B.Packets
    ' Fwd Packet Length Min',  # F.Packet Len Min
    ' Fwd Packet Length Mean',  # F.Packet Len Mean
    ' Bwd Packet Length Min',  # B.Packet Len Min
    ' Bwd Packet Length Std',  # B.Packet Len Std
    ' Flow IAT Mean',
    ' Flow IAT Std',
    ' Flow IAT Min',
    ' Fwd IAT Mean',  # FIAT Mean
    ' Fwd IAT Min',  # FIAT Min / Fwd IAT Min
    ' Bwd IAT Mean',
    ' Bwd IAT Min',
    'Fwd PSH Flags',  # FPSH Flags
    ' PSH Flag Count',
    ' ACK Flag Count',
    ' SYN Flag Count',
    'Fwd Packets/s',  # F.Packets/s
    ' Bwd Packets/s',  # B.Packets/s
    ' Subflow Fwd Bytes',  # Subflow F.Bytes
    ' Subflow Bwd Bytes',  # Subflow B.Bytes
    'Init_Win_bytes_forward',  # Init Win F.Bytes
    ' Init_Win_bytes_backward',  # Init Win B.Bytes
    'Active Mean',
    ' Active Min',
    ' Label'  # Assuming you need the label column for classification
]
```

```python
df = df[selected_headers]
df['Fwd Packets/s'] = df['Fwd Packets/s'].astype(int)
df[' Bwd Packets/s'] = df[' Bwd Packets/s'].astype(int)
df[' Bwd IAT Mean'] = df[' Bwd IAT Mean'].astype(int)
df[' Fwd IAT Mean'] = df[' Fwd IAT Mean'].astype(int)
df[' Flow IAT Std'] = df[' Flow IAT Std'].astype(int)
df[' Flow IAT Mean'] = df[' Flow IAT Mean'].astype(int)
df['Active Mean'] = df['Active Mean'].astype(int)
df[' Bwd Packet Length Std'] = df[' Bwd Packet Length Std'].astype(int)
df[' Fwd Packet Length Mean'] = df[' Fwd Packet Length Mean'].astype(int)

df
```

```python
from sklearn.preprocessing import QuantileTransformer

# Transform all features into the scale of [0,1]
numeric_features = df.dtypes[df.dtypes != 'object'].index
scaler = QuantileTransformer()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

```python
df.describe()
```

```python
# Multiply the feature values by 255 to transform them into the scale of [0,255]
df[numeric_features] = df[numeric_features].apply(
    lambda x: (x*255).astype(int))
```
